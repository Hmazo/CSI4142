```python
import pandas as pd
import numpy as np
```

```python
fact = pd.read_csv('fact.csv')
fact = fact.drop(columns=['Unnamed: 0','surrogate keys'])
fact = fact.rename(columns={'country_name': 'base_country'})
display(fact)
```

```python
fact_country = pd.read_csv('country_migration_fact_table.csv')
fact_country = fact_country.drop(columns=['Unnamed: 0','surrogate keys'])
display(fact_country)
```

```python
fact_industry = pd.read_csv('fact_industry_migration.csv')
fact_industry = fact_industry.drop(columns=['Unnamed: 0'])
fact_industry = fact_industry.rename(columns={'country_name': 'base_country'})
display(fact_industry)
```

```python

result = pd.merge(fact_country, fact, left_on=['base_country', 'year'], right_on=['base_country', 'year'], how='inner')

```

```python
display(result)
```

```python
result = pd.merge(result, fact_industry, on=['base_country', 'year','industry_name'], how='inner')
```

```python
display(result)
```

```python
result.isnull().sum()
```

```python
result['surrogate keys'] = range(1,len(result)+1)
```

```python
result.to_csv("fact_final.csv")
```
