```python
import pandas as pd
import numpy as np
```

```python
skill =  pd.read_csv('industry_dimension_skills.csv')
skill = skill.drop(columns=['Unnamed: 0','surrogate keys'])
display(skill)
```

```python
industry =  pd.read_csv('Industry_dimension.csv')
industry = industry.drop(columns=['Unnamed: 0','surrogate keys'])
display(industry)
```

```python
result = pd.merge(industry, skill, on='industry_name', how='left')
display(result)
```

```python
result['surrogate keys'] = range(1,len(result)+1)
```

```python
display(result)
```

```python
result= result.drop(columns=['top_skills'])
```

```python
result.to_csv("Industry_dimension_final.csv")
```
