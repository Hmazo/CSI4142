```python
import pandas as pd
```

```python


df =  pd.read_excel('public_use-industry-employment-growth.xlsx',  sheet_name="Growth from Industry Transition") 
```

```python
# Remove unnecessary columns
df = df.drop(columns=['country_code','wb_region','wb_income',	'isic_section_index','isic_section_name','industry_id'])
```

```python
print(df.dtypes)
```

```python
df
```

```python
# Clean up
df['country_name'] = df['country_name'].str.strip().str.lower()
df['industry_name'] = df['industry_name'].str.strip().str.lower()

#chane & to and in industry name
df['industry_name'] = df['industry_name'].str.replace('&','and')
```

```python
df = df.dropna()
# Remove duplicates
df = df.drop_duplicates()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['country_name'].unique()
```

```python
# Mean growth rate of each industry across all countries
industry_mean = df.groupby('industry_name').mean().reset_index()
display(industry_mean)

```

```python
# Median growth rate of each industry across all countries
industry_median = df.groupby('industry_name').median().reset_index()
display(industry_median)
```

```python

# Total growth rate of each industry across all countries
industry_total = df.groupby('industry_name').sum().reset_index()
display(industry_total)
```

```python
yearsint = [2015,2016,2017,2018,2019]
years= df[['growth_rate_2015','growth_rate_2016','growth_rate_2017','growth_rate_2018','growth_rate_2019']]
fact = pd.DataFrame()
for index, row in df.iterrows():
    for year, name in zip(yearsint, years.columns):
      temp = row
      temp["year"]= year
      temp["growth"]= years.iloc[index][name]
      
      fact = fact.append(temp, ignore_index=True)
```

```python
fact
```

```python
fact= fact.drop(columns= ['growth_rate_2015','growth_rate_2016','growth_rate_2017','growth_rate_2018','growth_rate_2019'])
```

```python
fact['surrogate keys'] = range(1,len(fact)+1)
```

```python
fact
```

```python
fact.to_csv("fact.csv")
```
