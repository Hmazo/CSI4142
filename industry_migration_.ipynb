### Change the 10K to original number of migration of each year


```python
df['industry_migration_2015'] = df['net_per_10K_2015']
df['industry_migration_2016'] = df['net_per_10K_2016'] 
df['industry_migration_2017'] = df['net_per_10K_2017']
df['industry_migration_2018'] = df['net_per_10K_2018']
df['industry_migration_2019'] = df['net_per_10K_2019']
df = df.drop(['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019'], axis=1)
df
```

#### Add the isic_group name to the industry migration 
Input the isic csv file

```python
df_isic = pd.read_csv('linkedin_to_isic_rev_4_industry_mapping_0.csv')
```

```python
df_isic.dtypes
```

Drop the columns that are not needed

```python
df_isic = df_isic.drop(columns=['isic_section_name','isic_division','isic_division_name','industry_sk','isic_section'])
```

### Tranform
Change the df_isic industry_name same as df industry_name

```python
df_isic['industry_name'] = df_isic['industry_name'].str.title()
df_isic[:10]
```

Merge the isic_group name to the industry migration

```python
result = pd.merge(df, df_isic, on='industry_name' , how='left')
result
```

Drop the null values

```python
result = result.dropna()
result
```

```python
# Clean up
result['country_name'] = result['country_name'].str.strip().str.lower()
result['industry_name'] = result['industry_name'].str.strip().str.lower()
result['wb_income'] = result['wb_income'].str.strip().str.lower()
result['wb_region'] = result['wb_region'].str.strip().str.lower()
result['isic_section_name'] = result['isic_section_name'].str.strip().str.lower()
result['Industry_group_name'] = result['Industry_group_name'].str.strip().str.lower()
result['isic_section_index'] = result['isic_section_index'].str.strip().str.lower()



#chane & to and in industry name
result['industry_name'] = result['industry_name'].str.replace('&','and')
result['wb_region'] = result['wb_region'].str.replace('&','and')
result['Industry_group_name'] = result['Industry_group_name'].str.replace('&','and')

```

```python
result
```

# **Fact Table :  We will now prepare the dataframe that will be used to create the fact table from the result dataframe**


```python
years = result[["industry_migration_2015","industry_migration_2016","industry_migration_2017",	"industry_migration_2018"	,"industry_migration_2019"]]

yearsint = [2015,2016,2017,2018,2019]
```

```python
fact = pd.DataFrame()
for index, row in result.iterrows():
    
    for year, name in zip(yearsint, years.columns):

      temp = row
      temp['year'] = year
      temp['industry_migration'] = years.loc[index, name]
      
      
      fact = fact.append(temp, ignore_index=True)

```

```python
fact= fact[['country_name','industry_name','year','industry_migration']]
display(fact)
```

```python
fact.to_csv('fact_industry_migration.csv', index=True)
```

# **Industry Dimension : we will now create the dataframe that will be used to create the industry dimension**

```python
grouped = result.groupby(['industry_id', 'industry_name'])[["industry_migration_2015", "industry_migration_2016", "industry_migration_2017", "industry_migration_2018", "industry_migration_2019"]].agg(['mean', 'median'])
grouped.columns = grouped.columns.map('_'.join)

grouped = grouped.reset_index()

display(grouped)
```

```python
industry = result.drop(columns=['country_code','country_name','wb_income','wb_region','industry_migration_2015','industry_migration_2016','industry_migration_2017','industry_migration_2018','industry_migration_2019'])
display(industry)
```

```python
# Remove duplicates
industry = industry.drop_duplicates().reset_index(drop=True)
```

```python
industry = industry.drop(columns=['industry_id'])
```

```python
industry_dimension = pd.merge(industry, grouped, on='industry_name' , how='left')
```

```python
industry_dimension
```

```python
industry_dimension['surrogate keys'] = range(1,len(industry_dimension)+1)
```
